<h1>Einführung in die Verwendung von  <em>jupyter notebooks</em> </h1> 

<p>
Das quelloffene Projekt <a href="https://jupyter.org/install"> <em>jupyter</em> </a>
bietet eine web-basierte interaktive Umgebung für wissenschaftliche Berech&shy;nungen
und Datenauswertung. Die zu Grunde liegenden Dateien sind sogenannte &bdquo;Notebooks&ldquo;, 
die sowohl formatierten Text als auch Programmcode u.&thinsp;a. in der Sprache <em>Python</em> 
und die entsprechende Programmausgabe in Form von Text und Grafiken in einem einzigen
Dokument vom Typ <em>.ipynb</em> speichern.<br> 
<br>
Vorhandene <em>Python</em>-Skripte sind direkt oder mit sehr wenig Aufwand unter
<em>jupyter</em> lauffähig. Die <em>jupyter</em>-Oberfläche ersetzt eine
Entwicklungsumgebung wie z.B. <em>IDLE</em>.
Zur Formatierung eingegebener Texte wird eine leicht zu erlernende Markup-Sprache 
verwendet, die auch Ausdrücke in LaTeX korrekt verarbeitet. Die Konversion vorhandener
Textbausteine ist also leicht zu bewerkstelligen. Notebooks können in verschiedenen
Formaten als komplette Dokumente, insb. auch <em>.pdf</em> und <em>.html</em>, 
exportiert und ggf. ausgedruckt werden. <br>
<br>
Zur <b>Anzeige und Bearbeitung</b> dient ein beliebiger Web-Browser, der sich mit einem
<em>jupyter</em>-Server verbindet. <br>
<br>
Ein <em>jupyter</em>-Server kann entweder lokal auf dem gleichen System wie der Browser
laufen oder auch auf einem über das Netzwerk angebundenen Rechner. 
Bei Verwendung eines Servers im Netzwerk oder auch in einer Compute-Cloud muss das Gerät
des Endnutzers nicht besonders leistungsfähig sein, sondern lediglich einen der gängigen
Web-Browser unterstützten. <br>

<p>
Eine <b>lokale Installation</b> auf dem System des Nutzers ist leicht möglich, wenn
eine <em>Python</em>-Installation vorhanden ist. Dazu  den <em>jupyter</em>-Server
mit dem installationswerkzeug <em>pip</em> installieren:
<pre><code>      > pip install jupyterlab </code></pre>
</p>
<p>
Dies funktioniert problemlos unter Linux, Mac&nbsp;OS oder auf Windows-Systemen
mit installierter <em>Python</em>-Umgebung.
<ul>
  <li> unter <b> Linux </b> ist <em>Pyhton</em> immer dabei,
	häufig allerdings noch in der älteren Version 2.x. <br>
	Mit Hilfe des Paket-Managers sollte eine Version 3.5
        oder höher installiert werden;
  </li>
  <li> für <b> Windows&nbsp;10 </b> wird die Umgebung 
  <a href="https://sourceforge.net/projects/winpython"> WinPython</a> empfohlen, die 
  auch ohne Installation <br>
  z.B. von einem externen Speicher läuft. WinPython bringt 
  einen eigenen <em>jupyter</em>-Server mit;  
  </li>
  <li> für <b> MacOS</b> wird 
  <a href="https://docs.anaconda.com/anaconda/install/mac-os/"> 
           Anaconda </a> empfohlen. 
  </li>
</ul> 
</p>

<p>
Auf dem eigenen System kann nun mittels Eingabe von
<pre><code>      > jupyter-lab </code></pre>
<p>
auf der Kommandozeile der Server gestartet werden. Normalerweise startet
dann ein Web-Browser, und es erscheint die <em>jupyter</em>-Oberfläche. 
Sollte dies nicht der Fall sein, wird eine URL angezeigt, die Sie in 
Ihren Web-Browser eingeben können, nachdem Sie ihn selbst gestartet haben. <br> 
<br>
Wenn Sie einen Server im Netzwerk verwenden, geben Sie die URL des Servers
in die Adresszeile des Browsers ein. In der Regel müssen Sie sich mit 
Benutzernamen und Kennwort anmelden. Danach startet auch in diesem Fall
die <em>jupyter</em>-Oberfläche. <br>
</p>

<h2> Los geht's !</h2>
<p>
  Starten Sie einen <em>jupyter</em>-Server bzw. verbinden Sie sich über Ihren
    Web-Browser mit dem Server der Fakultät. <br>
    
  Wenn die <em>jupyter</em>-Oberfläche in Ihrem Web-Browser erscheint, kann
  es direkt los gehen. Sie sehen eine Menüleiste, darunter links ein Fenster
  mit einer Dateiliste des Arbeitsverzeichnisses, und rechts ein großes 
  Fenster für die Anzeige von Text, Programm-Code und Programmausgaben.<br>
  <br>
  Links oben in der Menüleiste gibt es ein Symbol "&#8613; (Upload)" für das
  (Hoch-)Laden eines <em>jupyter</em>-Notebooks in den Arbeitsbereich des
  <em>jupyter</em>-Servers.

  Klicken Sie es an und wählen Sie im Datei-Browser eine der eben
  heruntergeladnen Dateien des Typs <em>.ipynb</em> aus.

  Das ausgewählte Notebook erscheint nun in der Dateiliste; starten Sie
  es mit einem Doppelklick.
</p>

<p>
Sie sehen nun den Inhalt des ausgewählten Notebooks im Hauptfenster
der Oberfläche. Folgen Sie den Hinweisen im angezeigten Text; die
Tutorials ist so ausgelegt, dass sie selbsterklärend sein sollten. 
</p>
