```python
import matplotlib.pyplot as plt
import numpy as np

t_series = np.arange(0.05, 0.16, 0.01)

n_data = [1, 2, 7, 11, 19, 21, 17, 10, 9, 4, 2]

dist = [[t_series[i]] * n for (i, n) in enumerate(n_data)]
dist = [round(n, 2) for bin in dist for n in bin]

mean = sum(dist) / len(dist)
sigma = np.sqrt(sum([((val - mean) ** 2 ) for val in dist]) / (len(dist) - 1))

print(mean, sigma)
plt.axvline(x=mean, color='red', label='<t>')
plt.axvline(x=mean+sigma, color='orange', label='m+s')
plt.axvline(x=mean-sigma, color='orange', label='m-s')


plt.hist(dist, range=(0, 0.2), bins=20, align='left')
plt.xlabel('t in s')
plt.ylabel('n')
plt.title('Messung einer Reaktionszeit')
plt.show()

```

Es ergibt sich eine diskrete Verteilung mit $\sigma\approx 0.021 s$ und einem Mittelwert von $\approx 0.10 s$
