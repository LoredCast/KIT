```python
import matplotlib.pyplot as plt
import numpy as np
GRAVITATION_ACC = 9.81 # das ist die Erdbeschleunigung g
```

```python
def position( t, v0, alpha ):
    return(
    [v0*t*np.cos(alpha),
    v0*t*np.sin(alpha) - ((GRAVITATION_ACC/2) * (t**2))] # hier k√∂nnen Sie die bahnkurve eintragen
    )
    
def line( beta, ax ):
    _, b = ax.get_xlim()
    ax.plot( [0,b], [0, -np.tan(beta)*b], label = 'Schiefe Ebene'  )
    ax.legend()
```

```python
beta = np.pi/3
alphas = [0, np.pi/4 - beta/2, beta, np.pi/4]
t = np.linspace(0, 4, 100)
v0 = 10
for alpha in alphas:
    fig, ax = plt.subplots( 1,1 )
    # damit plotten wir die bahnkurve
    ax.plot( *position(t, v0, alpha), label = 'Bahnkurve' )
    # damit plotten wir die schiefe ebene
    line( beta, ax )
```
