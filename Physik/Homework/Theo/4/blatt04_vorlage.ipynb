```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def solution( n, t, d, K, m ):
    exp_n = n/2 + 1
    K_n = (2 * K) / (n * m)
    arg = (-np.sqrt(K_n) * exp_n * t) + (d ** (exp_n))

    x = [arg_t ** (1 / exp_n) for arg_t in arg]

    return x
```

```python
K, m, d = 1, 1, 0.3
t = np.linspace(0, 0.2,10000000)
fig, ax = plt.subplots( 1,1  )
for n in [1, 3, 5, 11]:
    # damit plotten wir die bahnkurve
    ax.plot( t, solution(n, t, d, K, m), label = f'n = {n}' )
    ax.legend()
```
